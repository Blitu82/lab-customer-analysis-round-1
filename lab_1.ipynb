```python
#3.Rearrange the columns in the dataframe as needed.
```

```python
file1.columns
```

```python
# Change column name from 'st' to 'state'
file_1 = file1.rename({'st': "state"}, axis=1)
```

```python
file_1.columns
```

```python
file_2 = file2.rename({'st': "state"}, axis=1)
```

```python
#4.Rearrange the column order for file2, so it matches file1
file_2 = file_2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
```

```python
file_2.columns
```

```python
#Rearrange the column order for file3, so it matches file1 and file2
file_3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
```

```python
#Check that all columns are the same for the 3 files.
#file_1.columns == file_2.columns
file_2.columns == file_3.columns
```

```python
#5.Concatenate the three dataframes
```

```python
df = pd.concat([file_1, file_2, file_3])
```

```python
#6. Which columns are numerical? 
#'income', 'monthly premium auto' and 'total claim amount'
```

```python
df._get_numeric_data()
```

```python
#7. Which columns are categorical?
# customer, state, gender, education, customer lifetime value, number of open complaints, policy type, vehicle class
```

```python
df.select_dtypes(include=['object'])
```

```python
#8.1 Delete the column education and the number of open complaints from the dataframe
df = df.drop(['education', 'number of open complaints'], axis=1)
```

```python
#8.2 Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type

df['customer lifetime value'] = df['customer lifetime value'].astype(str).str.replace('%', "").astype(float) * 100
```

```python
df
```

```python
#8.3 Check for duplicate row in the data and remove if any
df = df.drop_duplicates()
```

```python
#8.4 Filter out the data for customers who have an income of 0 or less.
df_filtered = df[df.income > 0]
```
